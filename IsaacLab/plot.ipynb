```python
import pandas as pd
import numpy as np
```

```python
bc_df = pd.read_csv('bc_eval.csv')
```

```python
bc_df["BC_eval - Reward / Total reward (mean)"].mean()
```

```python
bp_df = pd.read_csv('behavioral_policy_eval.csv')
```

```python
import matplotlib.pyplot as plt

# Apply time-weighted exponential moving average (EMA)
ema_rewards = bc_df["BC_eval - Reward / Total reward (mean)"].ewm(span=30, adjust=False).mean()
ema_rewards_bp = bp_df["BC_eval - Reward / Total reward (mean)"][:100].ewm(span=30, adjust=False).mean()

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(ema_rewards_bp, label=r'$\pi_\beta$')
plt.plot(ema_rewards, label=r'$\pi$', linewidth=2)
plt.title('Behavior Cloning Evaluation Reward Over Time')
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Make a bar plot of the two means
import matplotlib.pyplot as plt

means = [bc_df["BC_eval - Reward / Total reward (mean)"].mean(), bp_df["BC_eval - Reward / Total reward (mean)"].mean()]
labels = ['Behavior Cloning', 'Behavioral Policy']
plt.bar(labels, means, color=['blue', 'orange'])
plt.ylabel('Mean Reward / Total Reward')
plt.title('Comparison of Mean Rewards')
plt.ylim(0, 1)
plt.grid(axis='y')
plt.show()
```
